```python
#importing matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
```

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
#cos wave
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
fig =plt.figure()
ax =fig.add_axes([0,0,1,1])
languages=["english","tamil","malayalam","telugu","hindi"]
people = [100,90,40,70,500]
ax.bar(languages,people) 
plt.xlabel("Languages")
plt.ylabel("Number of People") 
plt.show()           
```

```python
fig =plt.figure()
ax =fig.add_axes([0,0,1,1])
languages=["english","tamil","malayalam","telugu","hindi"]
people = [100,90,40,70,500]
ax.pie(people, labels=languages, autopct= "%1.1f%%") 
plt.show()           
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="g")
ax.scatter(x,z,color="b")
plt.show()
```

```python
fig3 = plt.figure
ax = plt.axes(projection="3d")
z = 20*np.random.random(100)
x = np.random.random(100) #np.sin(z)
y = np.random.random(100) #np.coz(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```
